```python
#wap to count vowels and consonants in a string
name ='prashant'
data=['a','e','i','o','u']
vowel=0
cons=0

for i in name:
  if i in data:
    vowel+=1
  else:
    cons+=1

print(f"Number of vowels in name {name} is {vowel}")
print(f"Number of consonantsin name {name} is {cons}")
```

```python
#wap to accept any character and check the entered character is upper case,lower case , digit,and spcecial symbol

char = input("Enter any character: ")

if len(char) != 1:
    print("Please enter only a single character.")
elif 'a' <= char <= 'z':
    print(f"The character '{char}' is a Lowercase Alphabet.")
elif 'A' <= char <= 'Z':
    print(f"The character '{char}' is an Uppercase Alphabet.")
elif '0' <= char <= '9':
    print(f"The character '{char}' is a Digit.")
else:
    print(f"The character '{char}' is a Special Symbol.")
```

```python
v=['a','e','i','o','u']
w = input("enter the world where we will search the vowels:")

found=[]
for i in w:
  if i in v:
    if i not in found:
      found.append(i)
print('found vowels=',found)
print('Unique vowels',len(found),'from the given word=',w)
```
