```python
#wap to aceept three ppr marks and calc total marks, percentage and check if percentage is greater than or eqqual to 60 then he/she is eligible for placement or not
marks = int (input("Enter marks:"))
marks2 = int (input("Enter marks:"))
marks3 = int (input("Enter marks:"))
total = marks + marks2 + marks3
print("Total marks:", total)
percentage = (total/300)*100
print("Percentage:", percentage)
if percentage >= 60:
  print("Eligible for placement")
else:
  print("Not eligible for placement")
```
